`Decision Tree Regression` is a non-linear regression technique used to predict continuous values by splitting the data into smaller subsets based on feature values. The model resembles a tree-like structure where decisions are made at each node, ultimately leading to a prediction at the leaves.  
### How Decision Tree Regression Works:
#### 1. Splitting the Data:  
The data is divided into subsets based on features that reduce the error (usually the variance). The algorithm looks for the best splits by checking different feature values. The feature and value that minimizes the error are selected for the split.

#### 2. Recursive Splitting:  
This process continues recursively, with each subset being split further. The splitting continues until a stopping criterion is met, such as reaching a maximum tree depth or having a minimum number of samples in each leaf.

#### 3. Prediction at Leaves:  
Once the data is split sufficiently, the tree stops, and predictions are made. The value predicted at each leaf is usually the average of the target values in that subset of the data.

## Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
y
```

## Training the Decision Tree Regression model on whole dataset.

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

## Why Feature Scaling in not Important in Decision Tree.
- The predictions from `Decision Tree Regression` model are resulting from `successive` splits of the data, through the different `nodes` of the Tree.
- And Therefore, there are not some equations like we have in previous models such as `SVR` or `Simple Linear Regression`.


## Predicting the new result

```python
regressor.predict([[6.5]])
```

`transform()` method is not used, because there is no `Feature Scaling`.

## Visualising the Decision Tree Regression results (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regressor)')
plt.xlabel('position salaries')
plt.ylabel('level')
plt.show()
```
