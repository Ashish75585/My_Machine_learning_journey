# Random Forest Regression



## Overview

Random Forest combines many decision trees to form an ensemble. Each tree is built using a random subset of the data and a random subset of features at each split. By doing this, Random Forest reduces the correlation between the individual trees and thus prevents overfitting.

Screenshot (43).png

## Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Random Forest Regression model on whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X,y)
```

`n_estimators` is the number of `trees` in the `forest`.

## Predicting the new Result.

```python
regressor.predict([[6.5]])
```

## Visualising the random forest Regression results

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
